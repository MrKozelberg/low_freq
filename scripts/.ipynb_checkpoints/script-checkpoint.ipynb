### Импорт пакетов

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = '../data/'
```

### Параметр накачки

```python
data1 = np.genfromtxt('%sdata1.txt'%path)

p_n = data1[:, 0]
p_iz = data1[:, 1]

A = p_n / 200
```

### Зависимость мощности излучения от параметра накачки

```python
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(1, 1, 1)

ax.scatter(A, p_iz - p_iz[-1], label='Эксперимент')

p1 = np.poly1d(np.polyfit(A, p_iz, 1))
ax.plot(A, p1(A), color='black',
        linestyle='--', label='Лин. аппроксим.')

ax.set_ylabel('Мощность излучения, мВт')
ax.set_xlabel('Параметр накачки')

ax.grid()
ax.legend()
```

```python
data2 = np.genfromtxt('data2.txt')

power = data2[:, 0]
freq = data2[:, 1]

A_2 = power / 200

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(1, 1, 1)

G = freq**2 / (A_2 - 1)

ax.plot(A_2, G)
ax.plot(A_2, np.ones(len(A_2))*G.mean(), color='black', linestyle='--', label='Среднее значение G')

ax.set_ylabel('Параметр G')
ax.set_xlabel('Параметр накачки')

ax.grid()
ax.legend()
```

```python

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(1, 1, 1)

ax.scatter(A_2, freq**2, label='Эксперимент')

ax.plot(A_2, (A_2-1)*G[A_2<1.6].mean(), color='black',
        linestyle='--', label='Теоретическая кривая')

ax.set_ylabel(r'Квадрат частоты, мГц${}^2$')
ax.set_xlabel('Параметр накачки')

ax.grid()
ax.legend()

fig.savefig('figures/freq_vs_pn.eps', bbox_inches='tight')
```
